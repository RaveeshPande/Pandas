Difference between: **loc[] and iloc[]**

**loc[]: This is used when we have column or row names in the argument hence can be used for specific case only if the number of column and rows are huge then we cannot name them all or write the names in the code.**

**iloc[]: This is used when we need to provide index which doesnt matter which column name we have the code iterates with index hence it is benificial for large chunk of data.**

**As per rule of indexing both row and column in iloc starts from 0.**

HOW TO **MERGE** A DATAFRAME:


To merge a dataframe we need to use a function as .merge()


Mostly we merge the data when 2 dataframe have same column with same values and we need to see the complete relationship of all the columns.

Without right on and left on this will not work:

data_3=pd.merge(data_1,data_2)
MergeError: No common columns to perform merge on. Merge options: left_on=None, right_on=None, left_index=False, right_index=False

**CORRECT SYNTAX:**

data_3=pd.merge(data_1,data_2,right_on='Y',left_on='B')

**Hence it is mendatory to have a common column in 2 data frame to get it merged.**

Here we have changed the data format to seee the merge function here we can see that the function is mapping the data.

That means if **Y column and B column both should have the same data.**
